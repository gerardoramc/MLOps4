```python
# Get the minimum number from the column "ModifiedDate"
min_date = df['ModifiedDate'].min()
df['ModifiedDate'] = (df['ModifiedDate'] - min_date).dt.days
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(["ModifiedDate"], axis=1), df["ModifiedDate"], test_size=0.2, random_state=42)
```

```python
# Preprocess the data

# X drop all columns except for CompanyName and SalesPerson
X_train = X_train[["CompanyName", "SalesPerson"]]
X_test = X_test[["CompanyName", "SalesPerson"]]

# Apply onehot encoding CompanyName and SalesPerson
X_train = pd.get_dummies(X_train, columns=["CompanyName", "SalesPerson"], drop_first=True)
X_test = pd.get_dummies(X_test, columns=["CompanyName", "SalesPerson"], drop_first=True)

# Align the columns of the test set with the training set
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
# Round the predictions to the nearest integer
y_pred = y_pred.round()
# Convert the predictions to integers
y_pred = y_pred.astype(int)
print(r2_score(y_test, y_pred))
```

```python
import pickle

file = open('model.pkl', 'wb')
pickle.dump(model, file)
file.close()
```
